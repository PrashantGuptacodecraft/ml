```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets._samples_generator import make_blobs
x,y=make_blobs(n_samples=50, centers=2, cluster_std=.60, random_state=0)
plt.scatter(x[:,0],x[:,1],c=y,s=50,cmap='winter')
```

```python
from sklearn.svm import SVC
model=SVC(kernel='linear',C=1)
model.fit(x,y)
```

```python
w=model.coef_[0]
```

```python
b=model.intercept_[0]
```

```python
w,b
```

```python
x_points=np.linspace(x[:,0].min(),x[:,0].max(),100)
```

```python

y_points=-(w[0]*x_points + b)/w[1]
```

```python
y_margin1=-(w[0]*x_points + b-1)/w[1]
y_margin2=-(w[0]*x_points + b+1)/w[1]
```

```python
plt.Figure(figsize=(10,6))
plt.scatter(x[:,0],x[:,1],c=y,s=50,cmap='coolwarm')
plt.plot(x_points,y_points,'k--',linewidth=2,label='Hyperplane')
```

```python
plt.plot(x_points,y_margin1,'k--',linewidth=1.5,label='Margins')
plt.plot(x_points,y_margin2,'k--',linewidth=1.5,)
```

```python
plt.scatter(model.support_vectors_[:,0],model.support_vectors_[:,1],s=100,facecolors='none',edgecolors='black', linewidths=2,label='Support Vectors')
```

```python
plt.title('SVM: Hyperplane and Margins & Support Vectors')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2') 
plt.legend()
plt.grid(True)
```
