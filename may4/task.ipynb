```python
# Perceptron - All in One Cell (AND Gate)

import numpy as np

# Activation function
def activation(x):
    return 1 if x >= 0 else 0

# Dataset (AND gate)
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([0, 0, 0, 1])

# Initialize weights & bias
w = np.zeros(2)
b = 0
learning_rate = 0.1
epochs = 10

# Training
for epoch in range(epochs):
    print(f"\nEpoch {epoch+1}")
    
    for i in range(len(X)):
        summation = np.dot(X[i], w) + b   # w.x + b
        y_pred = activation(summation)
        
        error = y[i] - y_pred
        
        # Update rule
        w = w + learning_rate * error * X[i]
        b = b + learning_rate * error
        
        print(f"Input: {X[i]} Output: {y_pred} Error: {error}")

# Final weights
print("\nFinal Weights:", w)
print("Bias:", b)

# Testing
print("\nTesting Results:")
for i in range(len(X)):
    summation = np.dot(X[i], w) + b
    y_pred = activation(summation)
    print(f"{X[i]} -> {y_pred}")
```
